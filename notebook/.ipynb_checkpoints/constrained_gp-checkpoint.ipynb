```python
from Constrained_GaussianProcess import ConstrainedGP

import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import norm
import scipy.stats
import scipy.linalg
from pyDOE import lhs

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

# 1D example
Let's consider a simple example, we want to learn $y=\Phi(\frac{x-0.5}{0.2})$ by Gaussian process satisfying different types of constraints. 

We consider the input space to be $\mathcal{D}=[0, 1]$, and we have training data $\{(x_i,f(x_i))\}_{i=1}^3$ for $x_1=0.25, x_2=0.5, x_3=0.75$.

The Gaussian process is approximated by finite-dimensional approach $Y_m(x)=\sum_{j=1}^my(t_j)\phi_j(x)$, so first we need to specify `m`.
Besides we need to specify the `constraints` imposed to this problem, now only 'Boundedness', 'Monotonicity', 'Convexity' or their combinations
are considered. Finally we need to give the input space `interval`.

```python
m = 30  # number of basis functions

# the constraints should be a dictionary specify the the constraints imposed to the GP, 
# it has three keys 'increasing', 'bounded' and 'convex'
constraints = {'increasing': True, 'bounded': [], 'convex': False}  

interval=[0,1]  # the input space

# Now we can define a object
Gp = ConstrainedGP(m, constraints=constraints, interval=interval)

# Training data
def f(x):
    return norm().cdf((x - 0.5) / 0.2)
x_train = np.array([0.25, 0.5, 0.75])
y_train = f(x_train)
```

The constraints and the training data enforce the posterior distribution to be truncated multivariate Gaussian(tmg), the MCMC methods are used to approximate the posterior distribution, so apart from training data, `method` (`HMC` or `Gibbs`), required number
of samples `n` and the burn in numbers `burn_in` should be specified when fit the data.
After fitting, we can get the conditional mean at test point.

```python
Gp.fit_gp(x_train, y_train, n=100, burn_in=100, method='HMC')

x_test = np.arange(0, 1 + 0.01, 0.01)
y_test = f(x_test)

y_pred = Gp.mean(x_test) # get the mean of posterior distribution at test point.
```

# 2D example
The number of basis functions and range of each interval should be specified respectively. Other parameters are same as 1D case.

```python
m = [10, 10]
constraint = {'increasing': False, 'bounded': [-np.pi, np.pi], 'convex': False}
interval=[[0,1],[0,1]]
Gp = ConstrainedGP(m, constraints=constraint, interval=interval)
method = 'HMC'

def f(x1, x2):
    return -0.5*(np.sin(9*x1)-np.cos(9*x2))

# the training data
x_train = lhs(2, 30)   # generate by Latin hypercube sampling
z_train = f(x_train[:, 0], x_train[:, 1])

Gp.fit_gp(x_train, z_train.flatten(), n=100, burn_in=100, method=method)

t = np.arange(0, 1, 0.01)
t1, t2 = np.meshgrid(t, t)
x_test = np.array([t1.flatten(), t2.flatten()]).T
z_test = f(t1, t2)

z_pred = Gp.mean(x_test)
```

Examples of Gaussian models satisfying different types of inequality constraints for interpolating the function
$x\mapsto\Phi(\frac{x-0.5}{0.2})$

```python
def title(constraint):
    increasing = constraint['increasing']
    convex = constraint['convex']
    bounded = len(constraint['bounded']) > 0

    if increasing and not convex and not bounded:
        title = 'Monotonicity'
    elif increasing and not convex and bounded:
        title = 'Monotonicity and boundedness'
    elif increasing and convex and not bounded:
        title = 'Monotonicity and convexity'
    elif increasing and convex and bounded:
        title = 'Monotonicity, boundedness and convexity'
    elif not increasing and not convex and bounded:
        title = 'Boundedness'
    elif not increasing and convex and not bounded:
        title = 'Convexity'
    elif not increasing and convex and bounded:
        title = 'Convexity and boundedness'
    else:
        title = 'Unconstrained'
    return title

def plot_fig1D(constraints, sampling_method, n=100, burn_in=100):    

    Gp = ConstrainedGP(m=30, constraints=constraints)
    rv = norm()

    def f(x):
        return rv.cdf((x - 0.5) / 0.2)
    x_train = np.array([0.25, 0.5, 0.75])
  
    y_train = f(x_train)
    samples = Gp.fit_gp(x_train, y_train, n=100, burn_in=100, method=sampling_method)

    t = np.arange(0, 1 + 0.01, 0.01)
    y_true = f(t)

    y_pred = Gp.mean(t)

    ci_l, ci_u = Gp.confidence_interval()

    fig, axis = plt.subplots()
    axis.plot(t, y_true, 'r', label="true function")
    axis.plot(t, y_pred, 'b', label="sample")
    axis.fill_between(t, ci_l, ci_u, color='lightgrey')
    axis.plot(x_train, y_train, 'ko', label="training points")
    axis.set_xlim(0., 1.)
    axis.set_ylim(0., 1.)
    axis.set_xlabel('x')
    axis.set_ylabel('y(x)')
    axis.legend(loc='upper left')
    axis.set_title(title(constraints)+',method: ' + sampling_method)

    plt.show()
```

```python
# boundedness constraint only
plot_fig1D({'increasing': False, 'bounded': [0, 1], 'convex': False}, 'HMC')
```

```python
# monotonicity constraint only
plot_fig1D({'increasing': True, 'bounded': [], 'convex': False}, 'HMC')
```

```python
# monotonicity and boundedness
plot_fig1D({'increasing': True, 'bounded': [0, 1], 'convex': False}, 'HMC')
```
